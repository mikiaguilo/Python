## Ejercicios clase

Ejercicio 1. Crea un módulo en Python que permita obtener de forma aleatoria notas enteras
entre 1 y 10. (utiliza la librería Random).

```python
import aleatorio
aleatorio.aleatorio(1,10)
```

Ejercicio 2.  Utilizando el módulo anterior, genera 30 notas aleatorias y guárdalas en un archivo Excel

```python
notas=[]
df=pd.DataFrame({"Notas":[]})
for i in range(0,30):
    notas.append(aleatorio.aleatorio(1,10))
df['Notas']=notas
df.to_excel("Notas de clase.xlsx")
```

Ejercicio 3. Crea un gráfico que muestre si hay más aprobados (notas ≥ 5) o suspensos (notas < 5).

```python
df['Aprovado']=[0]*len(df)
for i in range(0, len(df)):
    if df.loc[i,"Notas"]>=5:
        df.loc[i,"Aprovado"]=1
y=df["Aprovado"].value_counts()
plt.bar(y.index,y.values)
plt.ylim(0,30)
plt.xticks([0, 1], ["Suspendido", "Aprobado"])
plt.xlabel("Aprovado o suspenso")
plt.ylabel("Cantidad de estudiantes")
plt.title("Número de aprobados y suspendidos")
plt.show
```
